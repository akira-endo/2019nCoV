```r
library(repr)
options(repr.plot.width=9,repr.plot.height=6)
plotnb<-function(R,k){
    barplot(c(dnbinom(0:19,k,mu=R),NA,pnbinom(19,k,mu=R,lower.tail=F)),ylim=c(0,0.8),las=2)
    axis(side=1,at=1:22*1.2-0.5,label=c(0:19,"","20+"),tick=F)
    axis(side=2,at=seq(0,0.8,0.1),label=rep("",9),tck=1,lwd=0.5,lty=2)
    legend("topright", paste0("R = ",R, ", k = ",k),bty="n",cex=1.5)
}
```

```r
plotnb(7,0.2)
```

```r
plotnb(0.55,0.15)
```

```r
plotnb(1.1,0.15)
```
